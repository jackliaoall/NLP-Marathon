### Q1:
請學員思考為何不直接使用one-hot encoding的方式建立詞向量，而要有其他計數方法與推論方法的詞向量產生方式？

### Answer:
是容量的關係, 舉例來說用one-hot encoding方法, 有10000個字會產生10000維的向量
若用word2vec的方法, 可自行決定向量的維度, 可決定為100維的向量, 比起one-hot來說少很多容量

### Q2:
相較於計數手法的詞向量(ex: one-hot, 共現矩陣, PPMI)，word2vec的方法有何優點？

### Answer:
word2vec最大的優點是可以用GPU來進行平性運算加速運算
