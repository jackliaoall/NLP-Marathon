### 作業目的:了解Ensemble中的Blending與Stacking

### Question: 請描述Blending與Stacking的差異

Answer: 參考https://zhuanlan.zhihu.com/p/91659366

1.Blending方式和Stacking方式很類似,相比Stacking更簡單點,兩者區別是:

Blending是直接準備好一部分10%留出集只在留出集 上繼續預測,用不相交的數據訓練不同的Base Model ,將它們的輸出取(加權)平均。實現簡單,但對訓練數據利用少了。

2.Blending的優點是:比Stacking簡單,不會造成數據穿越(所謂數據創越,就比如訓練部分數據時候用了全局的統計特徵,導致模型效果過分的好) , generaliers和Istackers使用不同的數據,可以隨時添加其他模型到blender中。

3.缺點在於: Blending只使用了一部分數據集作為留出集進行驗證 ,而Stacking使用多折交叉驗證,比使用單-留出集更加穩健
