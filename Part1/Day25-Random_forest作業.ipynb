### 作業目的: 了解何謂Bias-Variance Tradeoff

### Q1: 一般模型誤差可拆為三部分，這三部分的誤差分別是？

Answer: Total error(整體誤差) = Bias(偏移量) + variance(變異) + random error(隨機誤差)

### Q2: 請解釋何謂Bias-Variance Tradeoff

Answer: 整體誤差由Bias和variance所組成, 隨機誤差基本上可忽略不計, 要降低整體誤差必須降低Bias和variance, 而兩者會互相影響, 如何在Bias和variance之間取得平衡, 使得整體誤差下降, 這就是Bias-Variance Tradeoff所要表達的
